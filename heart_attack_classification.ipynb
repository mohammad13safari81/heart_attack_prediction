```python
import plotly.express as px

def stacked_plot(df, column, target):
    """
    Function to generate a stacked bar chart between two variables using Plotly.
    """
    # Calculate the percentage of each category within the specified column
    temp_df = (df.groupby([column, target]).size() / df.groupby(column)[target].count()).reset_index()

    # Create a stacked bar chart
    fig = px.bar(temp_df, x=column, y=0, color=target, barmode='stack',
                 labels={'0': 'Heart attack Percentage'})

    # Customize the plot layout
    fig.update_layout(title=f"Stacked Bar Chart: {column} vs. {target}",
                      autosize=False,
                      width=400,
                      height=400,
                      xaxis_title=column,
                      yaxis_title='Heart attack Percentage')

    # Show the plot
    fig.show()


# Example usage:
# stacked_plot(df, 'category', 'churn')

```

```python
# Group by trtbps_update and output, and calculate counts
output_counts = df.groupby(["trtbps", "output"]).size().reset_index(name="counts")

# Create the countplot
fig = px.bar(
    data_frame=output_counts,
    x="trtbps",
    y="counts",
    color="output",
    barmode="group",  # Group bars side by side
    labels={"trtbps": "trtbps", "counts": "Count"},
    title="Heart Attack Distribution by trtbps"
)

# Show the plot
fig.show()
```

```python
def trtbps(t):
    if t>=90 and t <100:
        return 1
    elif t>=100 and t<110:
        return 2
    elif t>=110 and t<120:
        return 3
    elif t>=120 and t<130:
        return 4
    elif t>=130 and t<140:
        return 5
    elif t>=140 and t<150:
        return 6
    elif t>=150 and t<160:
        return 7
    elif t>=160 and t<170:
        return 8
    elif t>=170 and t<180:
        return 9
    elif t>=180 and t<=190:
        return 10
    elif t>=190 and t<=200:
        return 11

df['trtbps_update'] = df['trtbps'].apply(lambda x: trtbps(x))
```

```python
# Group by trtbps_update and output, and calculate counts
outputs_counts = df.groupby(["output", "trtbps_update"]).size().reset_index(name="counts")

# Create the countplot
fig = px.bar(
    data_frame=outputs_counts,
    x="trtbps_update",
    y="counts",
    color="output",
    barmode="group",  # Group bars side by side
    labels={"trtbps_update": "trtbps_update", "counts": "Count"},
    title="Heart Attack Distribution by trtbps_update"
)

# Show the plot
fig.show()
```
